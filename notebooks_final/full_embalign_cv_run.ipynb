# EmbAlign Full Production LOOCV Run

This notebook runs the full Leave-One-Out Cross-Validation pipeline to benchmark EmbAlign's lineage identity predictions using ground-truth labeled live-imaging data.

```python
# Standard Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import EmbAlign Modules
from aligner.config import PipelineConfig, AtlasStrategy, MatcherType, InitStrategy
from aligner.benchmarking import BenchmarkingSuite
from aligner.transformer import RigidTransformer
from aligner.runner import PipelineEvaluator
from aligner.plot_utils import *
%load_ext autoreload
%autoreload 2
```

### Read in Labeled Data


```python
PTS_CSV = "data/final_registered_all_pts.csv"
pts = pd.read_csv(PTS_CSV)
pts.head()
```

### Run LOOCV Pipeline

```python
# Initialize and configure pipeline
suite = BenchmarkingSuite(full_df=pts, transformer=RigidTransformer())
suite.add_config("EmbAlign_CV", PipelineConfig.v3_0_production())

# Run LOOCV sweep
frames_results, cells_diagnostics = suite.run_sweep(verbose=True)
```

```python
# Visualize alignment performance
plot_embryo_performance(frames_results)
```

```python
# Cache
frames_results.to_csv("data/aligned_frame_results_cv.csv", index = False)
cells_diagnostics.to_csv("data/aligned_cell_results_cv.csv", index = False)
```
